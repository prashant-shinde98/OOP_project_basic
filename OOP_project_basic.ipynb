```python
# OOP project of a ATM interface
class ATM:
  def __init__(self):
    self._database = {}
    self.__main_menu()


  def __main_menu(self):
    ans = int(input("""
    Hello!, WElcome to ATM
    Enter 1 : Register new card
    Enter 2 : Login
    """))

    if ans == 1:
      self.__register()

    elif ans == 2:
      self.__login()

    else:
      pass


  def __register(self):
    self.card_num = int(input('Enter card number'))
    self.pin = int(input('Enter pin'))
    self.balance = int(input('Enter the balance'))

    if self.card_num in self._database:
      print('Card Already exits')
      self.__login()

    else:
      self._database[self.card_num] = [self.pin, self.balance]
      print('New card stored succesefully')
      self.__login()


  def __login(self):
    int(input('Enter the card number'))
    int(input('ENter the pin'))

    if self._database[self.card_num][0] == self.pin:
      print('Succesfully loged in!! ')
      self._second_menu()

  def _second_menu(self):
    X = int(input("""
    Enter 1 : Check Balance
    Enter 2 : Withdraw cash
    Enter 3 : Change pin
    Enter 4 : Exit
    """))

    if X == 1:
      print('Your Balance is: ', (self._database[self.balance][1]))

    elif X == 2:
      a = int(input('Enter the amout'))
      if a <= self.balance:
        self.balance = self.balance - a
        print('New Balance is: ',  self.balance)

    elif X == 3:
      b = int(input('Enter the old pin: '))
      if b == self.pin:
        int(input('Enter the NEW Pin'))
        self.pin = b
        print('NEW pin succesfully updated')

      else:
        print('Enter correct pin')

    elif X == 4:
      self.__main_menu()

    else:
      print("Enter valid option")



a = ATM()
```
